```python
import matplotlib.pyplot as plt
```

```python
# importamos las imagenes de los mapas
mapa_2020 = plt.imread('../data/results/map_2020.png')
mapa_2021 = plt.imread('../data/results/map_2021.png')
mapa_2022 = plt.imread('../data/results/map_2022.png')

# creamos la figura
fig, ax = plt.subplots(2, 2, figsize=(25,25))

# graficamos los mapas
ax[0][0].imshow(mapa_2020)
ax[0][1].imshow(mapa_2021)
ax[1][0].imshow(mapa_2022)

# agregamos los titulos
ax[0][0].set_title('2020', fontsize=20)
ax[0][1].set_title('2021', fontsize=20)
ax[1][0].set_title('2022', fontsize=20)

# ocultamos los ejes
ax[0][0].axis('off')
ax[0][1].axis('off')
ax[1][0].axis('off')
ax[1][1].axis('off')

# reducimos el espacio entre los subplots
plt.subplots_adjust(wspace=0.01, hspace=0.01)
```
